```python
num = int(input("enter the number of rows:"))
row = 0
while row < num:
    star = row +1
    while star > 0:
        print("*", end = " ")
        star = star -1
    row = row +1
    print()
```

```python
d = {'k1' :"value" ,'k2' :"values" , 'k3':{"k12": "sudh","k13": "gafasd"}}
```

```python
def dic_to_list(d):
    l = []
    for i in d.values():
        if type(i) != dict:
            l.append(i)
        if type(i)== dict :
            for j in i.values():
                l.append(j)
    return l
```
